```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
data = data[['Close']]
print("\nUsing 'Close' column only:")
print(data.head())
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Define sequence length
sequence_length = 60
```

```python
X, y = [], []
for i in range(sequence_length, len(scaled_data)):
    X.append(scaled_data[i - sequence_length:i, 0])   # past 60 days
    y.append(scaled_data[i, 0])                       # next day's price

X, y = np.array(X), np.array(y)

# Reshape to (samples, time_steps, features)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))
print("Input shape:", X.shape)  # (num_samples, 60, 1)

# Split into training and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(60, 1)),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25, activation='relu'),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()
```

```python
data = pd.read_csv(io.BytesIO(uploaded['TSLA.csv']))
print(data.head())
data = data[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]

# Add some engineered features
data['MA7'] = data['Close'].rolling(window=7).mean()       # 7-day moving average
data['MA21'] = data['Close'].rolling(window=21).mean()     # 21-day moving average
data['Return'] = data['Close'].pct_change()                # Daily returns
#Drop NaNs created by rolling
data = data.dropna()
print("\nData with features:")
print(data.head())
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

sequence_length = 60
X, y = [], []

for i in range(sequence_length, len(scaled_data)):
    X.append(scaled_data[i - sequence_length:i])   # past 60 days of all 8 features
    y.append(scaled_data[i, 3])                    # next day's 'Close' price (4th column)

X, y = np.array(X), np.array(y)
print("Input shape:", X.shape)  # (samples, 60, 8)
```

```python
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25, activation='relu'),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()
```

```python
history = model.fit(X_train, y_train, batch_size=32, epochs=20, validation_split=0.1)
```

```python
predictions = model.predict(X_test)
# Only inverse transform the 'Close' column
close_scaler = MinMaxScaler()
close_scaler.min_, close_scaler.scale_ = scaler.min_[3], scaler.scale_[3]
predictions = (predictions - close_scaler.min_) / close_scaler.scale_

y_true = (y_test - close_scaler.min_) / close_scaler.scale_

plt.figure(figsize=(12, 6))
plt.plot(y_true, color='blue', label='Actual Price')
plt.plot(predictions, color='red', label='Predicted Price')
plt.title('Stock Price Prediction with LSTM (8 Features)')
plt.xlabel('Days')
plt.ylabel('Price')
plt.legend()
plt.show()
```
